```python
import pandas as pd
import pathlib
import glob
import os


```

```python

#concate all text
path = "/Users/pr158admin/Downloads/US_VitalStatistics/"

print(path)
all_files = glob.glob(os.path.join(path, "*.txt"))
all_files
overdose_merge = pd.concat((pd.read_csv(f ,delimiter ="\t") for f in all_files))
```

```python

overdose_merge.sample(10)
```

```python
#distinct values of columns
overdose_merge["Drug/Alcohol Induced Cause Code"].unique()
```

```python
#filter the drug induced deaths
drug_code = ['D1','D2','D3','D4']
overdose_df = overdose_merge[overdose_merge["Drug/Alcohol Induced Cause Code"].isin(drug_code)]
```

```python
overdose_df.sample(10)
```

```python
#write df to csv in the  repo

overdose_df.to_csv("overdose_df.csv", encoding="utf-8", index=False)
```
